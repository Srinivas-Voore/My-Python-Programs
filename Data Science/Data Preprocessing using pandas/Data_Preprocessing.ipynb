```python
# running .csv using pandas
import pandas as pd
c=pd.read_csv('Data.csv')
c
```

```python
c.isna().sum()
```

```python
d=c.dropna()
d
```

```python
e=c.iloc[:,1:3];
f=c.iloc[:,1:3].values;
print(e)
#two : => two dimension array
f
```

```python
#Imputer handles missing values and replaces with mean,median,more_frequency(mode)

from sklearn.preprocessing import Imputer
i=Imputer(missing_values='NaN',strategy='mean',axis=0)
i=i.fit(c.iloc[:,1:3])
c.iloc[:,1:3]=i.transform(c.iloc[:,1:3])
c
```

```python
from sklearn.preprocessing import Imputer
i=Imputer(missing_values=40.00000,strategy='most_frequent',axis=0)
i=i.fit(c.iloc[:,1:3])
c.iloc[:,1:3]=i.transform(c.iloc[:,1:3])
c
```

```python
#use label encoder to label strings into numbers so that machine can understand
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
c.iloc[:,0]=l.fit_transform(c.iloc[:,0])
c.iloc[:,-1]=l.fit_transform(c.iloc[:,-1])
c
```

#one Hot encoding

```python
from sklearn.preprocessing import OneHotEncoder
o=OneHotEncoder(categorical_features=[0]);
x=o.fit_transform(c).toarray()
x
```
